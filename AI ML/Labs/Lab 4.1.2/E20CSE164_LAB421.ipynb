````python
import pandas as pd
import numpy as np
````

````python
df = pd.read_csv('adult.csv', sep = ',')
df.head(10)
````

````python
df.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
                'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss',
                'hours-per-week', 'native_country', 'target']
df.head()
````

````python
df.shape
````

````python
df.columns
````

````python
df.replace(" ?", np.NaN, inplace=True)
df.isnull().sum()
````

````python
mode_workclass = df['workclass'].mode()
mode_occupation = df['occupation'].mode()
mode_native_country = df['native_country'].mode()
print("Mode for Workclass -->",mode_workclass)
print("Mode for Occupation -->", mode_occupation)
print("Mode for Native-Country -->", mode_native_country)
````

````python
# Handling all missing values

df['workclass'].replace(np.nan, 'Private', inplace=True)
df['occupation'].replace(np.nan, 'Prof-specialty', inplace=True)
df['native_country'].replace(np.nan, 'United-States', inplace=True)


df.isna().sum()
````

````python
pd.get_dummies(df, columns = ["target"], drop_first=True)
````

````python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
````

````python
X = df.iloc[:,0:-1]
X = pd.get_dummies(X, drop_first=True)
X
````

````python
Y=df.iloc[:,-1:]
Y.replace(' <=50K', 0,inplace=True)
Y.replace(' >50K', 1,inplace=True)
````

````python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.21, random_state=32)
````

````python

NBC = GaussianNB()
NBC.fit(X_train,Y_train)
````

````python


Y_pred=NBC.predict(X_test)
````

````python

from sklearn.metrics import confusion_matrix

print(confusion_matrix(Y_test,Y_pred))
tp, fn, fp, tn = confusion_matrix(Y_test,Y_pred).ravel()
print("Precision Score:", tp/(tp+fp))
print("Recall:", tp/(tp+fn))
print("F1 Score:", 2*(tp/(tp+fp))*(tp/(tp+fn))/((tp/(tp+fp))+(tp/(tp+fn))))
print("Accuracy:",(tp+tn)/(tp+tn+fp+fn))
````
