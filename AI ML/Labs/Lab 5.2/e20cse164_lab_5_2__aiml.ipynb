```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('CarPrice_Assignment.csv')
```

```python
df.head(10)
```

```python
print(df.dtypes)
```

```python
df.isna().sum()
```

```python
df=df.replace(to_replace="4wd",value="fwd")
```

```python
df=df.drop(['car_ID'], axis=1)
```

```python
categorical_features = list(df.select_dtypes(include=['object']).columns)
print(categorical_features)
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for i in categorical_features:
    df[i] = le.fit_transform(df[i])
    df[i] = le.fit_transform(df[i])
df
```

```python
X=df.iloc[:,:-1]
Y=df.iloc[:,-1]
print(X.columns,"\n")
print(Y.name)
```

```python
print(Y)
```

```python
Y=np.array(Y)

Y=Y.reshape(-1,1)
print(Y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2)
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python

from sklearn.svm import SVR
regressor = SVR(kernel = 'rbf')
regressor.fit(X, Y)
```

```python
svr = SVR()
svr.fit(x_train,y_train)
predict = svr.predict(x_test)
predict
```

```python
from sklearn.metrics import mean_squared_error , r2_score
print("R2 test",r2_score(y_test,predict))
print("MSE test",mean_squared_error(y_test,predict))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
lr_predict = lr.predict(x_test)
```

```python
lr_predict
```

```python
print(r2_score(y_test,lr_predict))
print(mean_squared_error(y_test,lr_predict))
```
